```python
sorted_movieset = movieset.sort_values(['vote_count','vote_average', 'popularity'], ascending=False)
```

```python
sorted_movieset.head(100)
```

```python
top100_sorted = sorted_movieset.head(100)
```

```python
top100_sorted.shape
```

```python
sns.jointplot(x='budget', y='revenue', data=top100_sorted, kind='reg')
```

```python
sns.pairplot(top100_sorted)
```

```python
g1 = sns.countplot(x='production_countries', data=top100_sorted)
g1.set_xticklabels(g1.get_xticklabels(), rotation=90)
```

```python
g0 = sns.boxplot(x='title', y='vote_average', data=top100_sorted)
g0.set_xticklabels(g0.get_xticklabels(), rotation=90)
```

```python
g2 = sns.stripplot(x='production_countries', y='revenue', data=top100_sorted, jitter=True)
g2.set_xticklabels(g2.get_xticklabels(), rotation=90)
```

```python
g3 = sns.swarmplot(x='production_countries', y='revenue', data=top100_sorted)
g3.set_xticklabels(g3.get_xticklabels(), rotation=90)
sns.violinplot(x='production_countries', y='revenue', data=top100_sorted)
```
